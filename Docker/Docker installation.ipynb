# Docker Installation

1. Search for Portainer in Docker and Run the following command on PowerShell:

docker run portainer/portainer-ce



2. Once portainer is downloaded, go to:
> __Getting started with Portainer CE__ and 

> Click on __Deploy Portainer__ 

> Click on __Set-up a new Portainer Server Installation__

> Click on __Docker Standalone__

> Click on __Install Portainer CE with Docker on WSL / Docker Desktop__

> Run the following command in PowerShell to create the volume that Portainer Server will use to store its database: __docker volume create portainer_data__

> Download and install the Portainer Server container by running the following command:
__docker run -p 8000:8000 -p 9443:9443 --name portainer -v /var/run/docker.sock:/var/run/docker.sock -v portainer_data:/data portainer/portainer-ce:latest__

Note that -d (detached) and --restart=always (restart option for container) have been removed


__Note__
You can remove a container as follows:

i. Stop the portainer by running this command: __docker container stop portainer__

ii. Then, remove it by running this command: ___docker ps -a -q | % { docker rm $_ }__

3. Login in to the localhpst interface by running the following address in a browser: 

__https://localhost:9443/#!/init/admin__
