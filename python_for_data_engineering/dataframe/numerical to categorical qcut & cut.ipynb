```python
# dividing into equal parts
df['Salary Category']=pd.qcut(df.Salary, 5, labels=False)
```

```python
# dividing numerical values into categorical using quartlile values
df['Salary Category']=pd.qcut(df.Salary, [0,.25,.75,1], labels=False)
df['Salary Category'].value_counts()
```

```python
df['Salary Category']
```

```python
# dividing numerical values into categorical: inclusive right boundary
df['Salary Category1']=pd.cut(df.Salary, [0,100000,150000], include_lowest=True)
df['Salary Category1'].value_counts()
```

```python
df['Salary Category1']
```

```python
# dividing numerical values into categorical: right boundary not included, left boundary not included
df['Salary Category2']=pd.cut(df.Salary, [0,100000,150000], right=False)
df['Salary Category2']
```

```python
# Proper labeling

bins = [0, 100000,150000]
labels = ["1. <100000","2. 100000-150000"]
df['Salary Category3']=pd.cut(df.Salary, bins=bins, labels=labels)
df['Salary Category3']
```

```python
cols=['Salary Category3','Salary']
df[cols]
```
