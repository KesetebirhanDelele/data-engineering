# Imputing Missing Values

```python
import pandas as pd
df=pd.read_csv('../data/employees.csv')
```

```python
df.info()
```

```python
# Map of data with missing values labled as 'True'
df.isnull()
```

```python
# Drop all records with at least one missing value
df2=df.dropna()
df2.shape
```

```python
df[df['Gender'].isnull()]
```

```python
df[df['Senior Management'].isnull()]
```

```python
df_categorical=df
df_categorical.info()
```

```python
# filling missing categorical variables with most common class
df_categorical_clean = df_categorical.apply(lambda x: x.fillna(x.value_counts().index[0]))
df_categorical_clean.info()
```

```python
df_unknown = df
# filling all categorical variables with Unknown class
df_unknown_clean = df_unknown.fillna("Unknown")
df_unknown_clean
```

```python
# filling all categorical variables with Unknown class
df_team = df_unknown['Team'].fillna("Unknown")
df_team
```

```python
# filling a specific categorical variable with Unknown class
df_unknown['Team'].fillna("Unknown",inplace=True)
df_unknown
```
