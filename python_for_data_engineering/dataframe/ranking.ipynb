```python
import pandas as pd
df=pd.read_csv('../data/employees.csv')
```

```python
df['default_rank'] = df['Salary'].rank()
df['max_rank'] = df['Salary'].rank(method='max')
df['NA_bottom'] = df['Salary'].rank(na_option='bottom')
df['pct_rank'] = df['Salary'].rank(pct=True)
cols=['Salary','default_rank','max_rank','NA_bottom','pct_rank']
df[cols]
```

```python
df2=df[cols]
df2
```
