# Reshaping Data: Merging (Appending Columns)

```python
import pandas as pd
```

```python
df=pd.read_csv('../data/Demo_sales_data.csv')
df.head()
```

```python
df1=df[['Row ID','Customer ID','Customer Name']]
df1.head()
```

```python
df2=df[['Segment','Country']]
df2.head()
```

```python
df3=pd.concat([df1,df2],axis=1)
df3.head()
```
