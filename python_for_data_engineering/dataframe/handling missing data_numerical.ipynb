```python
import pandas as pd
df=pd.read_csv('../data/sales.csv')
```

```python
df.info()
```

```python
df['quantity'].fillna(df['quantity'].min(),inplace=True)
df['price'].fillna(df['price'].max(),inplace=True)
df['bought'].fillna(df['bought'].median(),inplace=True)
df['beforenoon'].fillna(df['beforenoon'].mean(),inplace=True)
df['afternoon'].fillna(df['afternoon'].mode(),inplace=True)
```
