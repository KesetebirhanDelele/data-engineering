```python
import pandas as pd
df=pd.read_csv('../data/employees.csv')
```

```python
df.info()
```

```python
# Histogram for each numerical column
df.plot.hist()
```

```python
# Histogram for Salary
df.Salary.plot.hist()
```

```python
df.plot.scatter(x='Salary',y='Bonus %')
```

# Pie Chart: Option 1 

```python
df.value_counts('Gender').plot.pie(autopct='%1.1f%%', legend='Gender')
```

# Pie Chart: Option 2

```python
df.groupby(['Gender']).sum().plot(kind='pie', y='Salary', autopct='%1.1f%%')
```

```python
# Remove null values since the below code will not work
df_clean=df.dropna()
df_clean.info()
```

# Order X-axis by category and show value of bars

```python
ax=df_clean['Team'].value_counts().loc[df_clean['Team'].sort_values().unique()].plot.bar(
    xlabel='Team',ylabel='Frequency',title='Number of Employees by Team')

#annotate bars
ax.bar_label(ax.containers[0])
```
