```python
import pandas as pd
```

```python
# Create a DataFrame with dummy data.
df = pd.DataFrame(data={'staff_no': [9999] * 5,
                        'name': ['Dean McGrath'] * 5,
                        'year': [2016, 2017, 2018, 2019, 2020],
                        'hours': [349, 231, 876, 679, 976]})
df
```

```python
# Pivot
df2 = df.pivot(index=['staff_no', 'name'],
            columns=['year'], values='hours').reset_index()
df2
```

```python
# Unpivot the original DataFrame.
df3 = df2.melt(id_vars=['staff_no', 'name'], var_name='year',
             value_name='hours')
df3
```
