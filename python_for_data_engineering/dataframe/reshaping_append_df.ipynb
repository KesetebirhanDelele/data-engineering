# Shaping DataFrame: Append

```python
import pandas as pd
```

```python
df=pd.read_csv('../data/Demo_sales_data.csv')
df.head()
```

```python
df1=df.query("Segment=='Consumer'")
df1.head()
```

```python
df2=df.query("Segment=='Corporate'")
df2.head()
```

```python
# Appending: Set axis=0 or leave it empty
df3=pd.concat([df1,df2],axis=0)
df3
```
