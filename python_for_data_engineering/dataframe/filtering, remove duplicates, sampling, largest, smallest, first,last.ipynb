# Subsetting row values

## filtering, remove duplicates, sampling fraction or number, largest and smallest values, first and last values

```python
import pandas as pd
df=pd.read_csv('../data/Demo_sales_data.csv')
df.head()
```

```python
# Extract rows that meet logical criteria.
df[df['Row ID'] < 7]
```

```python
# Remove duplicate rows (only considers columns).
df.drop_duplicates()
```

```python
# Randomly select fraction of rows. 
df.sample(frac=0.001)
```

```python
# Randomly select n rows.
df.sample(n=4)
```

```python
# Select and order top n entries.
df.nlargest(5, 'Postal Code')
```

```python
# Select and order bottom n entries.
df.nsmallest(5, 'Postal Code')
```

```python
# Select first n rows.
df.head(5)
```

```python
# Select last n rows.
df.tail(5)
```
