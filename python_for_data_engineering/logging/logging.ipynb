```python
import logging
import pandas as pd
```

```python
# load a csv file
df = pd.read_csv(
    "../data/data_table_for_weekly_case_trends__the_united_states.csv", encoding='utf-8', nrows=100)
df.head()
```

```python
# create logger

# config the default level to debug - This debugs even the debug level
logging.basicConfig(
    filename="./reading_csvs.log",
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s",  # time level message
)
# > None

def is_positive(number):
    logging.debug(f"Checking if {number} is positive.")
    if number > 0:
        logging.info(f"{number} is positive?")
    else:
        logging.error(f"Error! Number {number} is zero!")


quantities = df["New Historic Cases"].to_list()

# write a error log
# write information to log file
for quantity in quantities:
    is_positive(quantity)
# example from the created reading_csvs.log file
# 2021-12-07 14:04:02,032 - INFO - 10 is positive?
# 2021-12-07 14: 04: 02, 032 - DEBUG - Checking if -1 is positive.
# 2021-12-07 14: 04: 02, 032 - ERROR - Error! Number - 1 is negative!
# 2021-12-07 14: 04: 02, 032 - DEBUG - Checking if 12 is positive.
```
