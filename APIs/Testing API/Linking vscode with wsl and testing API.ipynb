# Linking VSCODE with WSL

1. Click blue button at the left bottm corner


2. Select 'Connect to WSL using Distro' option from interpreter window


3. Select Ubuntu and that will connect you to the server

4. Click on interpreter or download Python as an extension 

5. To test API, open a new text file and save it as main.py using the content below:

        from fastapi import FastAPI

        app = FastAPI()


        @app.get("/")
        async def root():
            return {"message": "Hello World"}


6. Click 'Run the python file in new terminal' and paste and run the following:

        uvicorn main:app --reload

7. Open this in new browser and see the output 'Hello World':

        http://127.0.0.1:8000/
